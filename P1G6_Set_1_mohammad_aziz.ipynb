# **Conceptual Problems**

Jawab pertanyaan berikut:

1. Apakah yang dimaksud dengan inertia pada algoritma K-Means ?
2. Jelaskan yang dimaksud dengan Elbow Method (alasan penggunaan, cara penggunaan, kelemahan/kelebihan, dll) !

1. **Inertia pada algoritma K-Means:**
   Inertia pada algoritma K-Means adalah ukuran yang digunakan untuk mengevaluasi kualitas klasterisasi yang dihasilkan oleh algoritma K-Means. Inertia mengukur seberapa jauh titik-titik dalam sebuah klaster tersebar dari pusat klaster (centroid) yang bersangkutan. Inertia dihitung sebagai jumlah jarak kuadrat setiap titik dalam klaster terhadap pusat klasternya sendiri, dan kemudian jumlahkan untuk semua klaster. Dengan kata lain, inertia mencoba untuk mengukur seberapa padat (compact) klaster-klaster tersebut, dengan nilai inertia yang lebih rendah menunjukkan bahwa titik-titik dalam klaster cenderung berdekatan dengan centroidnya, yang mengindikasikan klasterisasi yang lebih baik.

2. **Elbow Method:**
   - **Alasan Penggunaan:** Elbow Method digunakan dalam konteks algoritma K-Means untuk menentukan jumlah optimal klaster yang harus digunakan dalam data. Hal ini karena ketika jumlah klaster meningkat, inertia umumnya akan menurun, karena klaster menjadi lebih kecil dan lebih padat. Namun, penurunan ini akan semakin berkurang seiring dengan penambahan klaster, hingga mencapai titik di mana penurunan ini tidak lagi signifikan, dan kurva inertia mulai membentuk "siku" (elbow) yang terlihat. Pada titik inilah jumlah klaster yang optimal dapat dipilih.
   - **Cara Penggunaan:** Cara penggunaan Elbow Method adalah dengan melakukan iterasi melalui berbagai nilai jumlah klaster (biasanya dari 1 hingga sejumlah tertentu) dan menghitung nilai inertia untuk setiap iterasi. Setelah itu, plot grafik antara jumlah klaster dan nilai inertia. Titik di mana grafik mulai membentuk "siku" adalah titik di mana jumlah klaster yang optimal dapat dipilih.
   - **Kelebihan:** Kelebihan dari Elbow Method adalah sederhana dan mudah untuk diinterpretasikan. Ini memberikan cara yang intuitif untuk menentukan jumlah klaster yang optimal dengan memperhatikan trade-off antara penurunan inertia dengan penambahan klaster.
   - **Kelemahan:** Salah satu kelemahan utama dari Elbow Method adalah tidak selalu mudah untuk menentukan titik "elbow" secara tepat dalam grafik inertia versus jumlah klaster. Terkadang grafik tersebut tidak memiliki siku yang jelas, atau memiliki beberapa kemungkinan titik elbow yang dapat dipilih. Selain itu, pada beberapa dataset yang kompleks, siku mungkin tidak terlihat dengan jelas, dan algoritma evaluasi lainnya seperti Silhouette Method dapat lebih diinginkan.
